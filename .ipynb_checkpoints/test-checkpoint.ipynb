## Présentation de MNIST (Modified National Institute of Standards and Technology)
La base de données MNIST pour Modified ou Mixed National Institute of Standards and Technology, est une base de données de chiffres écrits à la main. C'est un jeu de données très utilisé en apprentissage automatique.

### Formats de fichier
La base de données est composées de fichiers de format IDX3 et IDX1. Les propriétés des données sont stockées dans l'entête du fichier (header)<br>
IDX3 : format de fichier contenant des images
IDX1 : format de fichier contenant des classes

#### Header
header idx3 : valeur codée sur 16 bytes (magic_number;nb_image;rows;pixels)<br>
header idx1 : valeur codée sur 8 bytes (magic_number;nb_image)

```python
import struct
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def read_idx3(path):
    with open(path, "rb") as f:
        # Lecture du header
        magic, n_images, n_rows, n_cols = struct.unpack(">IIII", f.read(16))

        if magic != 0x00000803:
            raise ValueError(f"Magic number incorrect : {hex(magic)} (attendu : 0x00000803)")

        # Lecture des données brutes
        data = np.frombuffer(f.read(), dtype=np.uint8)

    # Conversion en tableau d'images
    return data.reshape((n_images, n_rows, n_cols))


def show_image(img):
    plt.imshow(img, cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
if __name__ == "__main__":
    images = read_idx3("MNIST/train-images.idx3-ubyte")

    print("Nombre d'images :", images.shape[0])
    print("Taille :", images.shape[1], "x", images.shape[2])

    # Afficher la première image
    show_image(images[0])
```
